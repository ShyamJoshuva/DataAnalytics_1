# analysis_1

The students who completed the test preparation course and those who didn't completed were taken and seen the performance in math,reading,writing score

variables used = {a,b,a_mean,b_mean,a_keys,a_values,b_keys,b_values}

```python
# number of unique values
df['test preparation course'].nunique()
```

```python
# unique values
df['test preparation course'].unique()
```

```python
# df - filered those who didnt take test preparation course
a = df[df['test preparation course']=='none']
a
```

```python
# df - filtered those who took or completed test preparation course 
b = df[df['test preparation course']=='completed']
b
```

```python
a.describe()
```

```python
b.describe()
```

```python
# dic variable which contains the mean values of various scores of students who completed test preparation course

a_mean = {'math_score':a['math score'].mean(),
          'reading_score':a['reading score'].mean(),
          'wrinting_score':a['writing score'].mean()}
a_mean.values()
```

```python
b_mean = {'math_score':b['math score'].mean(),
          'reading_score':b['reading score'].mean(),
          'writing_score':b['writing score'].mean()}
b_mean.values()
```

```python
a_keys = a_mean.keys()
a_values = a_mean.values()
plt.figure(figsize=(6,6))
plt.bar(a_keys,a_values,color ='orange',width=0.5)
#plt.xlabel('score')
#plt.ylabel('percentage')
plt.title("Students who didnt took-Test preparation course")
```

```python
b_keys = b_mean.keys()
b_values = b_mean.values()
plt.figure(figsize=(6,6))
plt.bar(b_keys,b_values,color ='blue',width=0.5)
#plt.xlabel('scores')
#plt.ylabel('percentage')
plt.title('Students who took-Test preparation course')
```

## results
* the students who completed the test preparation course(b) have a higher score in math, reading, writing score when compared to the students who didnt complete the course(a).
